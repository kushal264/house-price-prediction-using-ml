```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(housing,housing['CHAX']):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
housing=strat_train_set.drop('MEDV',axis=1)
housing_labels=strat_train_set['MEDV'].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(housing)
```

```python
imputer.statistics_
X=imputer.transform(housing)
housing_tr=pd.DataFrame(X,columns=housing.columns)
```

```python
housing_tr.describe()
```

```python
housing['TAXRM']=housing['TAX']/housing['RM']
housing['TAXRM'].head()
```

```python
housing.isnull().sum()
```

```python
housing.drop('TAXRM', axis=1, inplace=True)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([('imputer',SimpleImputer(strategy='median')),
                      ('std_scaler',StandardScaler()),])
```

```python
housing_num_tr=my_pipeline.fit_transform(housing_tr)
housing_num_tr.shape
```

```python
print(housing_num_tr.shape)      # e.g., (506, 13)
print(housing_labels.shape)      # e.g., (506,)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

##model = LinearRegression()
##model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
some_data=housing.iloc[:5]
```

```python
some_labels=housing_labels.iloc[:5]
```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
housing_predictions=model.predict(housing_num_tr)
mean_sq_error=mean_squared_error(housing_labels,housing_predictions)
mean_sq_error
```

```python
lin_rmse=np.sqrt(mean_sq_error)
```

```python
lin_rmse
```

evaluation cross validation

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,housing_num_tr,housing_labels,scoring="neg_mean_squared_error",cv=10)
rmse_score=np.sqrt(-scores)
```

```python
rmse_score
```

```python
def print_scores(scores):
    print("scores:",scores)
    print("mean:",scores.mean())
    print("stand. deviation",scores.std())
```

# Testing the model on test data


```python
# Testing the model on test data
from sklearn.metrics import mean_squared_error
import numpy as np

X_test = strat_test_set.drop("MEDV", axis=1)
Y_test = strat_test_set["MEDV"].copy()

X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)

final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

print("RMSE on test set:", final_rmse)
```
